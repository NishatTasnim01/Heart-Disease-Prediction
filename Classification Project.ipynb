```python
def calculate_accuracy_precision(confusion_matrix):
    tp, fn, fp, tn = confusion_matrix.ravel()
    accuracy = (tp + tn) / (tp + fp + fn + tn)
    precision = tp / (tp + fp)
    return accuracy, precision

dt_accuracy, dt_precision = calculate_accuracy_precision(dt_cm)
knn_accuracy, knn_precision = calculate_accuracy_precision(knn_cm)
rf_accuracy, rf_precision = calculate_accuracy_precision(rf_cm)
lr_accuracy, lr_precision = calculate_accuracy_precision(lr_cm)

print("Decision Tree Confusion Matrix Accuracy:", dt_accuracy)
print("Decision Tree Confusion Matrix Precision:", dt_precision)
print("K-Nearest Neighbors Confusion Matrix Accuracy:", knn_accuracy)
print("K-Nearest Neighbors Confusion Matrix Precision:", knn_precision)
print("Random Forest Confusion Matrix Accuracy:", rf_accuracy)
print("Random Forest Confusion Matrix Precision:", rf_precision)
print("Logistic Regression Confusion Matrix Accuracy:", lr_accuracy)
print("Logistic Regression Confusion Matrix Precision:", lr_precision)
```

## Handling Imbalanced Classes By Sampling Techniques

Sampling Techniques : 
1. Synthetic Minority Over-sampling Technique(SMOTE)
2. NearMiss for Under-Sampling
3. Over-Sampling

#### Synthetic Minority Over-sampling Technique(SMOTE)

```python
from imblearn.combine import SMOTETomek
sm = SMOTETomek(random_state=42)

SMOTE_X, SMOTE_Y = sm.fit_resample(x,y)
SMOTE_Y.value_counts()
```

#### NearMiss for Under-Sampling

```python
from imblearn.under_sampling import NearMiss
nm = NearMiss()

US_X, US_Y  = nm.fit_resample(x,y)
US_Y.value_counts()
```

#### Over-Sampling

```python
from imblearn.over_sampling import RandomOverSampler
over = RandomOverSampler(random_state=42)

OS_X, OS_Y = over.fit_resample(x,y)
OS_Y.value_counts()
```
